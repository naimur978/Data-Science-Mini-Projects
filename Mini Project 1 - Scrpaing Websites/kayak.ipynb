```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from time import sleep
from selenium_stealth import stealth
import re
import csv
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC

from selenium.common.exceptions import TimeoutException

chrome_options = webdriver.ChromeOptions()


# google colab allows only headless mode => no gui

# disable this locally if you want to see the browser window

chrome_options.add_argument('--headless')

# set window size, click() does not work if window has no size

chrome_options.add_argument('--window-size=1920,1080')

# other options, might not be needed

chrome_options.add_argument('--no-sandbox')

chrome_options.add_argument('--disable-dev-shm-usage')

chrome_options.add_argument('--enable-javascript')



driver= webdriver.Chrome(options=chrome_options)



stealth(driver,
        languages=["en-US", "en"],
        vendor="Google Inc.",
        platform="Win32",
        webgl_vendor="Intel Inc.",
        renderer="Intel Iris OpenGL Engine",
        fix_hairline=True,
        )
```

```python
def press_button_pagination(driver):
    try:
        more_button = WebDriverWait(driver, 10).until(
            EC.presence_of_element_located(
               (By.XPATH, "(//div[@class='ULvh-button show-more-button'])")
            )
        )
        more_button.click()
        WebDriverWait(driver, 30)
    except Exception as e:
        print(str(e))

def initialize_driver(url):
    driver.get(
        url,
    )
    for _ in range(10):
        try:
          WebDriverWait(driver, 10).until(
              EC.presence_of_element_located(
                (By.XPATH, "(//div[@class='ULvh-button show-more-button'])")
              )
          )
          press_button_pagination(driver=driver)
        except Exception as e:
            print(str(e))

    webpage = driver.page_source
    return driver, webpage

initialize_driver("https://www.kayak.com/flights/HEL-PAR/2023-10-10?sort=bestflight_a")
```

```python
import re

def ignore_plus_one(string):
  """Ignores the +1 from a string like "8:50 am +1".

  Args:
    string: The string to be modified.

  Returns:
    A modified string without the +1.
  """

  pattern = re.compile(r'\+1$')
  string = re.sub(pattern, '', string)
  return string
```

```python
arrival_times = []
flights = driver.find_elements(By.XPATH, "(//div[@class='Fxw9']//div[@class='vmXl vmXl-mod-variant-large'])")

print(len(flights))
# print(len(flights))
for flight in flights:
    arrival_time = ignore_plus_one(flight.text).split(' – ')

    arrival_times.append(arrival_time)
# print(arrival_times)
```

```python
airlines = []
flights = driver.find_elements(By.XPATH, "(//div[@class='Fxw9']//div[@class='VY2U']//div[@class='c_cgF c_cgF-mod-variant-default'])")

print(len(flights))
for flight in flights:
    arrival_time = flight.text
    # print(arrival_time)
    airlines.append(arrival_time)
# print(airlines)
```

```python
import re

def show_integer(string):
  """Shows the integer in a string, or prints "nonstop" if there is no integer.

  Args:
    string: The string to be processed.

  Returns:
    The integer in the string, or "nonstop" if there is no integer.
  """

  pattern = re.compile(r'\d+')
  match = pattern.search(string)

  if match:
    return int(match.group())
  else:
    return 0
```

```python
stops = []
flights = driver.find_elements(By.XPATH,"(//div[@class='Fxw9']//div[@class='JWEO'])")

print(len(flights))
for flight in flights:
    stop = flight.text
    stop = show_integer(stop)
    stops.append(stop)
# print(stops)
```

```python
prices = []
flights = driver.find_elements(By.XPATH,
"(//div[@class='Fxw9']//div[@class='f8F1-price-text'])")

print(len(flights))
for flight in flights:
    stop = flight.text
    stop = re.compile(r'\$').sub('', stop).replace(',', '')
    prices.append(stop)
# print(prices)
```

```python
duration = []
flights = driver.find_elements(By.XPATH, "(//div[@class='Fxw9']//div[@class='xdW8']//div[@class='vmXl vmXl-mod-variant-default'])")

print(len(flights))
for flight in flights:
    stop = flight.text
    duration.append(stop)
# print(duration)
```

```python
def remove_after_m(string):
  """Removes everything after the first "m" in a string.

  Args:
    string: The string to be processed.

  Returns:
    The string with everything after the first "m" removed.
  """

  index = string.find("m")
  if index != -1:
    return string[:index+1]
  else:
    return string


```

```python
layover = []
layover_city = []


flights = driver.find_elements(By.XPATH, "//div[@class='Fxw9']//div[@class='nrc6']")

print(len(flights))

for flight in flights:
    class_name = flight.text

    # print(flight.get_attribute('data-resultid'))

    variable_name = flight.get_attribute('data-resultid')
    xpath = f"//*[@data-resultid='{variable_name}']//div[@class = 'nrc6-wrapper']//div[@class = 'JWEO']"

    flight2 = flight.find_element(By.XPATH, xpath)

    flight3 = flight2.find_element(By.XPATH, ".//div[@class = 'c_cgF c_cgF-mod-variant-default']")




    all_attributes = flight3.get_attribute('outerHTML')

    # print(all_attributes)


    if 'title' in all_attributes:
        span_index = all_attributes.find("title")
        # Assuming that the attribute you want to extract is 'your_attribute'
        span_attribute_start = all_attributes.find("title=\"", span_index)
        span_attribute_end = all_attributes.find("\"", span_attribute_start + len("title=\""))
        span_attribute_value = all_attributes[span_attribute_start + len("title=\""):span_attribute_end]

        duration2 = remove_after_m(span_attribute_value)
        # print(duration2)
        # print(flight3.text)

        layover.append(duration2)
        layover_city.append(flight3.text)
    else:
        layover.append('none')
        layover_city.append('none')
        # print('none')
# print(layover_city)
```
